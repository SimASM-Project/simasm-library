# Warehouse Outbound Process - Stutter Equivalence Verification

This notebook demonstrates verification of a complex **multi-station warehouse outbound process** modeled in both Event Graph and ACD formalisms.

## Warehouse Process Overview

The warehouse has 6 service stations:
- **S1**: Pick Zone A (4 servers)
- **S2**: Pick Zone B (2 servers)
- **S3**: Label (3 servers)
- **S4**: Scan (1 server)
- **S5**: Pack (3 servers)
- **S6**: Release (2 servers)

Jobs enter through 4 creation streams (2 for Zone A, 2 for Zone B), then flow through Label → Scan → Pack → Release.

## Observation Level W

We verify equivalence using:

**W = { s1_busy_eq_*, s2_busy_eq_*, s3_busy_eq_*, s4_busy_eq_*, s5_busy_eq_*, s6_busy_eq_* }**

This tracks the number of busy servers at each station.

```python
# Entity Flow Trace Analysis
# ==========================
# This section traces entities through the warehouse to verify correct flow
# from creation through all stations to departure.

# Re-register models for trace collection
simasm.register_model("warehouse_eg", eg_simasm_code)
simasm.register_model("warehouse_acd", acd_simasm_code)

# Run EG model and collect state trace
from simasm.parser import load_string
from simasm.runtime.stepper import ASMStepper, StepperConfig

def collect_state_trace(model_source: str, seed: int = 42, end_time: float = 10000.0):
    """Run model and collect state variables at each step."""
    loaded = load_string(model_source, seed=seed)
    
    main_rule = loaded.rules.get(loaded.main_rule_name)
    config = StepperConfig(time_var="sim_clocktime", end_time=end_time)
    stepper = ASMStepper(
        state=loaded.state,
        main_rule=main_rule,
        rule_evaluator=loaded.rule_evaluator,
        config=config,
    )
    
    # State variables to track
    trace_data = {
        'time': [],
        'Q1': [], 'Q2': [], 'Q3': [], 'Q4': [], 'Q5': [], 'Q6': [],
        'S1_busy': [], 'S2_busy': [], 'S3_busy': [], 'S4_busy': [], 'S5_busy': [], 'S6_busy': [],
        'arrivals': [],      # Cumulative arrivals (Zone A + Zone B)
        'arrivals_A': [],    # Cumulative arrivals to Zone A
        'arrivals_B': [],    # Cumulative arrivals to Zone B
        'departures': []
    }
    
    # Track cumulative arrivals by detecting Q1/Q2 increases
    cumulative_arrivals = 0
    cumulative_arrivals_A = 0
    cumulative_arrivals_B = 0
    prev_Q1 = 0
    prev_Q2 = 0
    prev_S1_busy = 0
    prev_S2_busy = 0
    
    def record_state():
        nonlocal cumulative_arrivals, cumulative_arrivals_A, cumulative_arrivals_B
        nonlocal prev_Q1, prev_Q2, prev_S1_busy, prev_S2_busy
        
        trace_data['time'].append(stepper.sim_time)
        
        # Queue lengths
        Q1 = loaded.state.get_var('Q1')
        Q2 = loaded.state.get_var('Q2')
        trace_data['Q1'].append(Q1)
        trace_data['Q2'].append(Q2)
        trace_data['Q3'].append(loaded.state.get_var('Q3'))
        trace_data['Q4'].append(loaded.state.get_var('Q4'))
        trace_data['Q5'].append(loaded.state.get_var('Q5'))
        trace_data['Q6'].append(loaded.state.get_var('Q6'))
        
        # Busy servers = capacity - available
        s1_cap = loaded.state.get_var('S1_capacity')
        s1_avail = loaded.state.get_var('S1')
        S1_busy = s1_cap - s1_avail
        trace_data['S1_busy'].append(S1_busy)
        
        s2_cap = loaded.state.get_var('S2_capacity')
        s2_avail = loaded.state.get_var('S2')
        S2_busy = s2_cap - s2_avail
        trace_data['S2_busy'].append(S2_busy)
        
        s3_cap = loaded.state.get_var('S3_capacity')
        s3_avail = loaded.state.get_var('S3')
        trace_data['S3_busy'].append(s3_cap - s3_avail)
        s4_cap = loaded.state.get_var('S4_capacity')
        s4_avail = loaded.state.get_var('S4')
        trace_data['S4_busy'].append(s4_cap - s4_avail)
        s5_cap = loaded.state.get_var('S5_capacity')
        s5_avail = loaded.state.get_var('S5')
        trace_data['S5_busy'].append(s5_cap - s5_avail)
        s6_cap = loaded.state.get_var('S6_capacity')
        s6_avail = loaded.state.get_var('S6')
        trace_data['S6_busy'].append(s6_cap - s6_avail)
        
        # Detect arrivals: job arrived if (Q + busy) increased at entry stations
        zone_A_total = Q1 + S1_busy
        zone_B_total = Q2 + S2_busy
        prev_zone_A_total = prev_Q1 + prev_S1_busy
        prev_zone_B_total = prev_Q2 + prev_S2_busy
        
        if zone_A_total > prev_zone_A_total:
            cumulative_arrivals_A += (zone_A_total - prev_zone_A_total)
            cumulative_arrivals += (zone_A_total - prev_zone_A_total)
        if zone_B_total > prev_zone_B_total:
            cumulative_arrivals_B += (zone_B_total - prev_zone_B_total)
            cumulative_arrivals += (zone_B_total - prev_zone_B_total)
        
        trace_data['arrivals'].append(cumulative_arrivals)
        trace_data['arrivals_A'].append(cumulative_arrivals_A)
        trace_data['arrivals_B'].append(cumulative_arrivals_B)
        trace_data['departures'].append(loaded.state.get_var('departures'))
        
        # Update previous values
        prev_Q1 = Q1
        prev_Q2 = Q2
        prev_S1_busy = S1_busy
        prev_S2_busy = S2_busy
    
    record_state()  # Initial state
    
    # Run simulation
    while stepper.sim_time < end_time:
        if not stepper.step():
            break
        record_state()
    
    return trace_data

print("Collecting EG trace (seed=42)...")
eg_trace = collect_state_trace(eg_simasm_code, seed=42, end_time=10000.0)
print(f"  Collected {len(eg_trace['time'])} state snapshots")
print(f"  Final time: {eg_trace['time'][-1]:.2f}")
print(f"  Total arrivals: {eg_trace['arrivals'][-1]} (Zone A: {eg_trace['arrivals_A'][-1]}, Zone B: {eg_trace['arrivals_B'][-1]})")
print(f"  Total departures: {eg_trace['departures'][-1]}")
```

```python
# Compute time-weighted averages for entity flow statistics

def compute_time_weighted_average(times, values):
    """Compute time-weighted average of a trace."""
    if len(times) < 2:
        return np.mean(values)
    
    # Time intervals
    dt = np.diff(times)
    # Use value at start of each interval
    weighted_sum = np.sum(np.array(values[:-1]) * dt)
    total_time = times[-1] - times[0]
    return weighted_sum / total_time if total_time > 0 else 0

# Station configuration
stations = [
    ('S1', 'Pick Zone A', 4, 'Q1', 'S1_busy'),
    ('S2', 'Pick Zone B', 2, 'Q2', 'S2_busy'),
    ('S3', 'Label', 3, 'Q3', 'S3_busy'),
    ('S4', 'Scan', 1, 'Q4', 'S4_busy'),
    ('S5', 'Pack', 3, 'Q5', 'S5_busy'),
    ('S6', 'Release', 2, 'Q6', 'S6_busy'),
]

# Compute statistics
print("=" * 80)
print("ENTITY FLOW STATISTICS (Event Graph Model, seed=42)")
print("=" * 80)
print(f"\n{'Station':<12} {'Name':<14} {'Cap':>4} {'Avg Queue':>10} {'Max Queue':>10} {'Util %':>8}")
print("-" * 80)

for station_id, name, capacity, q_key, s_key in stations:
    avg_queue = compute_time_weighted_average(eg_trace['time'], eg_trace[q_key])
    max_queue = max(eg_trace[q_key])
    avg_busy = compute_time_weighted_average(eg_trace['time'], eg_trace[s_key])
    utilization = (avg_busy / capacity) * 100
    
    print(f"{station_id:<12} {name:<14} {capacity:>4} {avg_queue:>10.2f} {max_queue:>10} {utilization:>7.1f}%")

print("-" * 80)
print(f"\nTotal Departures: {eg_trace['departures'][-1]} jobs")
print(f"Simulation Time: {eg_trace['time'][-1]:.2f}")
print(f"Throughput: {eg_trace['departures'][-1] / eg_trace['time'][-1] * 1000:.3f} jobs per 1000 time units")
```

```python
# Visualize station utilization as bar chart
fig, ax = plt.subplots(figsize=(12, 5))

station_names = []
utilizations = []
capacities = []

for station_id, name, capacity, q_key, s_key in stations:
    avg_busy = compute_time_weighted_average(eg_trace['time'], eg_trace[s_key])
    utilization = (avg_busy / capacity) * 100
    station_names.append(f"{station_id}\n{name}\n(cap={capacity})")
    utilizations.append(utilization)
    capacities.append(capacity)

x = np.arange(len(station_names))
bars = ax.bar(x, utilizations, color=['#3498db', '#2ecc71', '#9b59b6', '#e74c3c', '#f39c12', '#1abc9c'])

# Add utilization percentage labels on bars
for bar, util in zip(bars, utilizations):
    height = bar.get_height()
    ax.annotate(f'{util:.1f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom', fontsize=11, fontweight='bold')

ax.set_ylabel('Utilization (%)', fontsize=12)
ax.set_title('Station Utilization (% of Capacity Used)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(station_names)
ax.set_ylim(0, 100)
ax.axhline(y=80, color='r', linestyle='--', alpha=0.5, label='80% threshold')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Identify bottleneck
max_util_idx = np.argmax(utilizations)
print(f"\nBottleneck Analysis:")
print(f"   Highest utilization: {stations[max_util_idx][0]} ({stations[max_util_idx][1]}) at {utilizations[max_util_idx]:.1f}%")
print(f"   This station is most likely to cause congestion.")
```

```python
# Flow Conservation Check
# =======================
# For a stable queueing system: Arrivals = Departures + WIP

# Count total jobs in system (WIP = queues + in-service)
final_wip = (
    eg_trace['Q1'][-1] + eg_trace['Q2'][-1] + eg_trace['Q3'][-1] + 
    eg_trace['Q4'][-1] + eg_trace['Q5'][-1] + eg_trace['Q6'][-1] +
    eg_trace['S1_busy'][-1] + eg_trace['S2_busy'][-1] + eg_trace['S3_busy'][-1] +
    eg_trace['S4_busy'][-1] + eg_trace['S5_busy'][-1] + eg_trace['S6_busy'][-1]
)

arrivals = eg_trace['arrivals'][-1]
departures = eg_trace['departures'][-1]
expected_wip = arrivals - departures

print("=" * 60)
print("FLOW CONSERVATION CHECK")
print("=" * 60)
print(f"Total Arrivals:                  {arrivals}")
print(f"  - Zone A arrivals:             {eg_trace['arrivals_A'][-1]}")
print(f"  - Zone B arrivals:             {eg_trace['arrivals_B'][-1]}")
print(f"Departures (completed jobs):     {departures}")
print(f"Work-in-Progress (WIP):          {final_wip}")
print(f"Expected WIP (Arrivals - Dep):   {expected_wip}")
print("-" * 60)
if final_wip == expected_wip:
    print("FLOW CONSERVATION: VERIFIED")
    print("Arrivals = Departures + WIP")
else:
    print(f"WARNING: Mismatch! Difference = {final_wip - expected_wip}")
print("=" * 60)
```
